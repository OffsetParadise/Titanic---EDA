## Analysis 

```python
## Count of Survived vs Not survived

sns.countplot(x="Survived", data = Data)
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number Of Passengers")
plt.show()
```

```python
## Gender Distriburtion

sns.countplot(x="Sex", data=Data)
plt.title("Gender Distribution")
plt.show()
```

```python
## Passenger CLass Distriburtion

sns.countplot(x="Pclass", data=Data)
plt.title("Passenger Class Distribution")
plt.show()
```

```python
## Survival Rate by Gender


sns.countplot(x="Survived", hue="Sex", data=Data)
plt.title("Survival By Class")
plt.show()
```

```python
## Survival By Passenger Class

sns.countplot(x="Survived", hue="Pclass", data=Data)
plt.title("Survival By Class")
plt.show()
```

```python
## Survival By Embarked Location


sns.countplot(x="Survived", hue="Embarked", data=Data)
plt.title("Survival By Embarkation Port")
plt.show()
```

```python
## Age Distribution by Survival 


sns.kdeplot(data=Data[Data['Survived'] == 1]['Age'],
            label='Survived', shade=True)
sns.kdeplot(data=Data[Data['Survived'] == 0]['Age'],
            label='Did not Survive', shade=True)
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.legend()
plt.show()
```

## Correlation Heatmap

```python
Data_corr = Data.copy()
```

```python
Data_corr["Sex"] = Data_corr["Sex"].map({"male": 0, "female": 1})
```

```python
Data_corr["Embarked"] = Data_corr["Embarked"].map({"s": 0, "C": 1, "Q": 2})
```

```python
Data_corr = Data_corr.drop(["Name"], axis =1)
```

```python
Data_corr = Data_corr.dropna()
```

```python
plt.figure(figsize = (10, 6))
sns.heatmap(Data_corr.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
